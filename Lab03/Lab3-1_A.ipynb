**2.2 Suppose the cover price of a book is $24.95, but bookstores get a 40% discount. Shipping costs $3 for the first copy and 75 cents for each additional copy. What is the total wholesale cost for 60 copies?**
<br>
<br>
**Answer:**

```python
cover_price = 24.95
discount = 0.40
copies = 60

total_book_cost = cover_price * (1 - discount) * copies

shipping = 3 + 0.75 * (copies - 1)

total = total_book_cost + shipping

print(f"Total cost of the 60 books with applied discount: ${round(total_book_cost,2)}")
print(f"Ant the total paid including shipping costs: ${round(total,2)}")
```

**2.3 If I leave my house at 6:52 am and run 1 mile at an easy pace (8:15 per mile), then 3 miles at tempo (7:12 per mile) and 1 mile at easy pace again, what time do I get home for breakfast?**
<br>
<br>
**Answer:**

```python
start = 6 * 60 + 52
easy = 8 + 15/60
tempo = 7 + 12/60

# total run in easy pace is 2 miles, and total run in tempo pace is 3 miles, therefore:
total_run = (2 * easy) + (3 * tempo)
arrival = start + total_run

# Now we transform it into readable time format:
# // is called floor division and rounds down to a whole number, and % gives the remainder after the division (in this case the seconds)
hours = int(arrival//60)
minutes = int(arrival % 60)
seconds = round((arrival % 1) * 60)

print(f"{hours}:{minutes:02d}:{seconds:02d} AM") 
```
