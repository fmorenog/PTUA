```python
# 2. Suppose the cover price of a book is $24.95, but bookstores get a 40% discount.
# Shipping costs $3 for the first copy and 75 cents for each additional copy. What is the total wholesale cost for 60 copies?

book_price = 24.95
discount = 0.40
initial_shipping = 3
additional_shipping = 0.75
total_copies = 60

# calculate discount price per book
book_price_discount = book_price * (1 - discount)
print(f"Cover price per book: ${book_price_discount:.2f}")


total_cost = round(book_price_discount * total_copies, 2)
print(f"Total book cost without shipping: ${total_cost:.2f}")

# Calculate shipping cost

shipping_cost = initial_shipping + (total_copies - 1) * additional_shipping
print(f"Total shipping cost: ${shipping_cost:.2f}")


# Calculate wholesale cost
wholesale_cost = total_cost + shipping_cost
print(f"Total wholesale cost for 60 copies: ${wholesale_cost:.2f}")

```

3. If I leave my house at 6:52 am and run 1 mile at an easy pace (8:15 per mile), then 3 miles at tempo (7:12 per mile) and 1 mile at easy pace again, what time do I get home for breakfast?

```python
#import libraries
from datetime import datetime, timedelta

# calculate start time

start_time = datetime.strptime("6:52 AM", "%I:%M %p")
print(f"Run time begins at: {start_time.strftime('%I:%M %p')}")


# 1 mile at easy pace (8:15 per mile)
start_pace_min = 8
start_pace_sec = 15
start_pace_miles = 1

# 3 miles at 7:12 per mile
middle_pace_min = 7
middle_pace_sec = 12
middle_pace_miles = 3

# 1 mile at easy pace again (8:15 per mile)
last_pace_miles = 1


# Calculating pace for each pace segment
segment1_pace = timedelta(
    minutes = start_pace_min * start_pace_miles,
    seconds = start_pace_sec * start_pace_miles
)

segment2_pace = timedelta(
    minutes = middle_pace_min * middle_pace_miles,
    seconds = middle_pace_sec * middle_pace_miles
)

segment3_pace = timedelta(
    minutes = start_pace_min * last_pace_miles,
    seconds = start_pace_sec * last_pace_miles
)


# Calculate total run time
total_run_time = segment1_pace + segment2_pace + segment3_pace

print(f"Total run time: {total_run_time}")

# Calculate arrival time
arrival_time = start_time + total_run_time

print(f"Arrival time for breakfast: {arrival_time.strftime('%I:%M %p')}")
```
